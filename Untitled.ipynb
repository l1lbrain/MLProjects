```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
# load file csv
```

```python
data
# longitude: đo khoảng cách căn nhà về phía Đông, càng xa phía Đông thì giá càng cao
# latitude: đo khoảng cách căn nhà về phía Bắc, càng xa phía Bắc thì giá càng cao
# housing_median_age: tuổi trung bình của một căn nhà trong một dãy nhà, số càng thấp thì nhà càng mới
# total_rooms: tổng số phòng trong một dãy nhà
# total_bedrooms: tổng số phòng ngủ trong một dãy nhà
# population: tổng số người dân cư trú trong một dãy nhà
# households: tổng số hộ gia đình
# median_income: thu nhập trung bình của các hộ gia đình (đơn vị chục nghìn USD)
# median_house_value: giá trị trung bình của căn nhà (đơn vị USD)
# ocean_proximity: vị trí của căn nhà so với biển
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['median_house_value'], axis=1)
Y = data['median_house_value']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
train_data = X_train.join(Y_train)
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
from sklearn.linear_model import LinearRegression

X_train, Y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']

reg = LinearRegression()

reg.fit(X_train, Y_train)
```

```python
test_data = X_test.join(Y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity).astype(int)).drop(['ocean_proximity'], axis=1)
```

```python
X_test, Y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
reg.score(X_test, Y_test)
```

```python
reg.predict(X_test)
```
